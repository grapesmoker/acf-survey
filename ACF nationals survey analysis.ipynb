```python
import numpy as np
import pandas as pd
import os
from matplotlib import pyplot as plt
```

```python
source = 'acf-survey.csv'
data = pd.read_csv(source)
```

```python
transportation = data.dropna(subset=['transportation'])
transportation.transportation.value_counts().plot(kind='bar')
plt.ylabel('Number of schools')
```

```python
travel_time = np.array(data.dropna(subset=['travel_time']).travel_time)
fig = plt.hist(travel_time)
ax = plt.gca()
plt.xlabel('Travel time in hours')
plt.ylabel('Number of schools')
```

```python
travel_cost = np.array(data.dropna(subset=['travel_cost']).travel_cost)
fig = plt.hist(travel_cost)
ax = plt.gca()
plt.xlabel('Travel cost in $')
plt.ylabel('Number of schools')
```

```python
lodging_cost = np.array(data.dropna(subset=['lodging_cost']).lodging_cost)
fig = plt.hist(travel_cost)
ax = plt.gca()
plt.xlabel('Lodging cost in $')
plt.ylabel('Number of schools')
```

```python
total_cost = lodging_cost + travel_cost
fig = plt.hist(total_cost)
plt.xlabel('Total cost in $')
plt.ylabel('Number of schools')
```

```python
columns = ['acf_NE', 'acf_MidAtl', 'acf_SE', 'acf_EMidW', 'acf_CMidW', 'acf_NMidW', 'acf_WMidW', 'acf_MtnW', 'acf_Wcoast', 'acf_PNW']
descriptions = ['Northeast', 'Mid Atlantic', 'Southeast', 'Eastern Midwest', 'Central Midwest', 'North Midwest',
                'Western Midwest', 'Mountain West', 'West Coast', 'Pacific Northwest']
```

```python
willing = []
unwilling = []
indices = np.arange(len(columns))
width = 0.35

for column in columns:
    column_data = data[column].value_counts(sort=False)
    unwilling.append(column_data[0])
    willing.append(column_data[1])
    
plt.bar(indices, willing, width, color='g')
plt.bar(indices + width, unwilling, width, color='r')
ax = plt.gca()
ax.set_xticks(indices + width / 2)
labels = ax.set_xticklabels(descriptions, rotation='vertical')
plt.ylabel('Number of schools')
```
